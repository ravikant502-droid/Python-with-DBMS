```python
num = int(input("Enter a positive number: "))

count = 0
for i in range(1, num + 1):
    if num % i == 0:
        count += 1

print("Number of factors:", count)
```

```python
#Given 2 positive numbers, find GCD or HCF of both the numbers.

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

hcf = 1
for i in range(1, min(a , b) +1 ):
    if a % i == 0 and b % i == 0:
        hcf = i
        
print("HCF: ", hcf)
```

```python
#Given 2 positive numbers, find LCM of both the numbers.
#LCM
a = int(input("Enter first number"))
b = int(input("Enter secoand number"))

minValue = max(a , b)
maxValue = a * b
lcm = 0

for i in range(minValue, maxValue + 1):
    if(i % a == 0) and (i % b == 0):
        lcm = i
        break
print(lcm)
```
