```python
#star square
n = int(input())
for i in range(n): 
   for j in range(n):
       print("*", end=" ")
   print()
```

```python
# star right triangle
n = int(input())

i = 1
while i <= n:
    j = 1
    while j <= i:
        print(j, end = " ")
        j += 1
    i += 1
    print()
```

```python

n = int(input())

i = 1
while i <= n:
    j = 1
    while (j <= n - i + 1):
        print("*", end = " ")
        j += 1
    i += 1
    print()
```
