```python
#Given an array of integers arr[], the task is to find the maximum and minimum elements in the array using the minimum number of comparisons.
def findArr(arr):
    if len(arr) == 0:
        return None

    maxele = arr[0]
    minele = arr[0]

    for i in range(len(arr)):
        if arr[i] > maxele:
            maxele = arr[i]
        elif arr[i] < minele:
            minele = arr[i]

    return (maxele, minele)


# Example usage
arr = [1, 7, 2, 4, 10, 12, 15]
ans = findArr(arr)
print(ans)


        
        
```

```python
#Sort 0s, 1s and 2s: Given an array arr[] containing only 0s, 1s, and 2s. Sort the array in ascending order. Note: You need to solve this problem without utilizing the built-in sort function.[Dutch national flag]
def sort012(arr):
    start, mid = 0, 0
    end = len(arr) - 1

    while mid <= end:
        if arr[mid] == 0:
            arr[start], arr[mid] = arr[mid], arr[start]
            start += 1
            mid += 1
        elif arr[mid] == 1:
            mid += 1
        else:  # arr[mid] == 2
            arr[mid], arr[end] = arr[end], arr[mid]
            end -= 1

    return arr


# Example usage
arr = [0, 2, 1, 2, 0, 1, 0, 2, 1]
print("Sorted array:", sort012(arr))

```

```python
#Move all negative numbers to beginning and positive to end with constant extra space.
#Input: -12, 11, -13, -5, 6, -7, 5, -3, -6 
#Output: -12 -13 -5 -7 -3 -6 11 6 5
def movenums(arr):
    if len(arr) == 0:
        return arr

    start = 0
    end = len(arr) - 1

    while start <= end:
        if arr[start] < 0:
            start += 1
        elif arr[end] >= 0:
            end -= 1
        else:
            # Swap negative from end with positive at start
            arr[start], arr[end] = arr[end], arr[start]
            start += 1
            end -= 1

    return arr
arr = [-12, 11, -13, -5, 6, -7, 5, -3, -6] 
print("Rearranged array:", movenums(arr))

# Example usage


```

```python
#Given an array of length N and an integer x, you need to find if x is present in the array or not. Return true or false
def find(arr, x):
    # base case
    if len(arr) == 0:
        return False

    # check first element
    if arr[0] == x:
        return True

    # recursive call on remaining list
    return find(arr[1:], x)


l = [1, 2, 3, 4, 5, 6, 7, 8]
k = 6

findAns = find(l, k)
print(findAns)
```
