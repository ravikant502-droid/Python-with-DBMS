```python
#WAP to check or print the following
#Palindrome string

n = input("Enter a string")
if n == n[::-1]:
    print("the number is palindrome")
else:
    print("the number is not palindrome")
```

```python
#Fibonacci
n = int(input())
num1 = 0
num2 = 1
print(num1, end =" ")
print(num2, end =" ")
s = 0
for i in range (3, n+1):
    s = num1 + num2
    print(s, end= " ")
    #swap n1 
    num1 = num2
    num2 = s
```

```python
#factorial
# Factorial using for loop
n = int(input("Enter a number: "))
fact = 1

for i in range(1, n + 1):
    fact *= i

print(f"Factorial of {n} is {fact}")
```

```python
# WAP to count the occurrence of characters in a string

# declare empty dictionary
char_count = {}

# input string
string = input("Enter a string: ")

# traverse the string
for char in string:
    # check if character is present in dictionary
    if char in char_count:
        char_count[char] += 1
    else:
        # create entry starting from 1
        char_count[char] = 1

# print result
print("Character occurrences:")
print(char_count)
```
